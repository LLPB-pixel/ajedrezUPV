```python
import tkinter as tk
import chess
import chess.svg
from PIL import Image, ImageTk
import io
import os
import chess.pgn
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers
import chess.engine
import random
from tqdm import tqdm

# Cargar el modelo guardado
final_model_filepath = r"C:\Users\marco\OneDrive\Ajedrez Computacional\modelo_ajedrez_final.h5"
modelo_cargado = keras.models.load_model(final_model_filepath)

# Función para convertir un tablero de ajedrez a un array numpy
def board_to_array(board):
    array = np.zeros((8, 8, 12), dtype=np.int8)
    piece_map = board.piece_map()
    for square, piece in piece_map.items():
        piece_type = piece.piece_type - 1
        color = int(piece.color)
        array[square // 8, square % 8, piece_type + (6 * color)] = 1
    return array

# Clase para la aplicación de ajedrez con interfaz gráfica
class ChessApp:
    def __init__(self, root):
        self.root = root
        self.board = chess.Board()
        self.label = tk.Label(root)
        self.label.pack()
        self.update_board()

        self.entry = tk.Entry(root)
        self.entry.pack()
        self.button = tk.Button(root, text="Mover", command=self.make_move)
        self.button.pack()

    # Actualizar el tablero en la interfaz gráfica
    def update_board(self):
        svg_board = chess.svg.board(board=self.board)
        img = Image.open(io.BytesIO(svg_board.encode('utf-8')))
        photo = ImageTk.PhotoImage(img)
        self.label.configure(image=photo)
        self.label.image = photo

    # Realizar un movimiento en el tablero
    def make_move(self):
        move = self.entry.get()
        try:
            chess_move = chess.Move.from_uci(move)
            if chess_move in self.board.legal_moves:
                self.board.push(chess_move)
                self.update_board()
                self.bot_move()
            else:
                print("Movimiento ilegal")
        except:
            print("Formato de movimiento inválido")

    # Realizar el movimiento del bot
    def bot_move(self):
        posicion_array = board_to_array(self.board).reshape(1, 8, 8, 12)
        evaluaciones = modelo_cargado.predict(posicion_array)
        mejores_jugadas = sorted(self.board.legal_moves, key=lambda move: evaluaciones[0], reverse=True)
        self.board.push(mejores_jugadas[0])
        self.update_board()

# Iniciar la interfaz gráfica
root = tk.Tk()
app = ChessApp(root)
root.mainloop()
```
